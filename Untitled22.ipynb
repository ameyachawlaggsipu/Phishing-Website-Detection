```python
import pandas as pd
```

```python
data = pd.read_csv('masterdata.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
y = data['Label']
x = data.drop(['Domain','Label'],axis=1)
```

```python
print(x.shape,y.shape,data.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

clf = DecisionTreeClassifier(max_depth = 19)
clf.fit(train_x, train_y)
    
```

```python
from sklearn.model_selection import KFold
from sklearn.utils import shuffle
x,y=shuffle(x,y)
```

```python
kf = KFold(n_splits=15)
import numpy as np
x = np.array(x)
y = np.array(y)
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import PrecisionRecallDisplay
```

```python
from sklearn.metrics import confusion_matrix
model_storage = []
matrix = []
i = 1
for train_index, test_index in kf.split(x,y):
   #print(train_index,test_index)
    train_x, test_x = x[train_index], x[test_index]
    train_y, test_y = y[train_index], y[test_index]
    clf = DecisionTreeClassifier(max_depth = 19)
    clf.fit(train_x, train_y)
    r = clf.score(test_x,test_y)
    model_storage.append(clf)
    print(r)
    display = PrecisionRecallDisplay.from_estimator(clf, test_x, test_y, name="DecisionTree")
    _ = display.ax_.set_title("2-class Precision-Recall curve for model ")
    plt.show()
    i= i+1
    plot_confusion_matrix(clf, test_x, test_y)
    plt.show()
    matrix.append(confusion_matrix(test_y, clf.predict(test_x)))

```

```python
model_storage
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
eclf2 = VotingClassifier(estimators=[('dt0', model_storage[0]),('dt1', model_storage[1]),('dt2', model_storage[2]),('dt3', model_storage[3]),('dt4', model_storage[4]),('dt5', model_storage[5]),('dt6', model_storage[6]),('dt7', model_storage[7]),('dt8', model_storage[8]),('dt9', model_storage[9]),('dt10', model_storage[10]),],voting='hard')
```

```python
eclf2.fit(x,y)
```
